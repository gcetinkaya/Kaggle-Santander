## Calculates cosine similarity between trainLevel1Preds and testLevel1Preds
### We can use this data to choose the data set to train our fnn stacking model!


```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity as cossim
```

```python
trainDataFrame = pd.read_csv('./data/trainLevel1Preds.csv')
trainFeatures = trainDataFrame.drop(['TARGET'], axis=1)
```

```python
testDataFrame = pd.read_csv('./data/testLevel1Preds.csv')
testFeatures = testDataFrame.drop(['ID'], axis=1)
```

```python
print trainFeatures.shape
print testFeatures.shape
```

```python
trn_ixs = trainFeatures.index
trn_sims = []
for ix in trn_ixs:
    row = trainFeatures.ix[ix].reshape(1,-1)
    sims = cossim(row, testFeatures)
    sims_mean = np.mean(sims, axis=1)
    trn_sims.append(sims_mean)
    print "%5d...%.4f" % (ix, sims_mean)
```

```python
# save trn_sims
import joblib

joblib.dump(trn_sims, './data/trainLevel1Preds_cosims')
```
