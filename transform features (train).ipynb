```python
import joblib
import pandas as pd
import numpy as np

import sys
sys.path.insert(0, '/Users/gokhan/libs')

from Evio.FeatureGenerator import FeatureGenerator
```

```python
schema = joblib.load('./models/features/merged_schema_5000_features.joblib')
```

```python
# load original features
trainDataFrame = pd.read_csv('./data/train.csv')

# remove constant columns
colsToRemove1 = []
for col in trainDataFrame.columns:
    if trainDataFrame[col].std() == 0:
        colsToRemove1.append(col)

trainDataFrame.drop(colsToRemove1, axis=1, inplace=True)

# remove duplicate columns
colsToRemove2 = []
columns = trainDataFrame.columns
for i in range(len(columns)-1):
    v = trainDataFrame[columns[i]].values
    for j in range(i+1,len(columns)):
        if np.array_equal(v,trainDataFrame[columns[j]].values):
            colsToRemove2.append(columns[j])

trainDataFrame.drop(colsToRemove2, axis=1, inplace=True)
```

```python
trainLabels = trainDataFrame['TARGET']
trainFeatures = trainDataFrame.drop(['ID','TARGET'], axis=1)
```

```python
fg = FeatureGenerator()
trainTransformed = fg.generate_features_from_schema(trainFeatures, schema)
```

```python
trainTransformed['TARGET'] = trainLabels
```

```python
# remove constant columns
colsToRemove1 = []
for col in trainTransformed.columns:
    if trainTransformed[col].std() == 0:
        colsToRemove1.append(col)

# remove duplicate columns
colsToRemove2 = []
columns = trainTransformed.columns
for i in range(len(columns)-1):
    v = trainTransformed[columns[i]].values
    for j in range(i+1,len(columns)):
        if np.array_equal(v,trainTransformed[columns[j]].values):
            colsToRemove2.append(columns[j])
```

```python
print "num of columns w/out variance: %d" % len(colsToRemove1)
print "num of duplicate columns: %d" % len(set(colsToRemove2))
```

```python
# drop useless features
trainTransformed.drop(colsToRemove1, axis=1, inplace=True)
trainTransformed.drop(colsToRemove2, axis=1, inplace=True)
```
